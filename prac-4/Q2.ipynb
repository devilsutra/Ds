```python
class UndoEditor:
    def __init__(self):
        self.stack = [] 

    def insert(self, text):
        self.stack.append(text)
        print(f"Inserted: '{text}'")
        self.show()

    def undo(self):
        if not self.stack:
            print("Nothing to undo!")
            return
        else:
            removed = self.stack.pop()
            print(f"Undo: Removed '{removed}'")
            self.show()

    def show(self):
        print("Current Text:", "".join(self.stack))

editor = UndoEditor()
editor.insert('H')
editor.insert('e')
editor.insert('l')
editor.insert('l')
editor.insert('o') 

editor.undo() 
editor.undo()
editor.show()
```
