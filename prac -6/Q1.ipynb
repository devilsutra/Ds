```python
class Node:
    def __init__(self, roll):
        self.roll = roll
        self.left = None
        self.right = None

def insert(root, roll):
    if root is None:
        return Node(roll)
    if roll < root.roll:
        root.left = insert(root.left, roll)
    elif roll > root.roll:
        root.right = insert(root.right, roll)
   
    return root 

def inorder(root):
    return inorder(root.left) + [root.roll] + inorder(root.right) if root else []

def preorder(root):
    return [root.roll] + preorder(root.left) + preorder(root.right) if root else []

def postorder(root):
    return postorder(root.left) + postorder(root.right) + [root.roll] if root else []

# Test the implementation
roll_numbers = [105, 101, 110, 103, 107]
root = None
for roll in roll_numbers:
    root = insert(root, roll)

print("In-order traversal (sorted):", inorder(root))
print("Pre-order traversal:", preorder(root))
print("Post-order traversal:", postorder(root))
```
