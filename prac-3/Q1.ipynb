```python
class SongNode:
    def __init__(self, name):
        self.name = name
        self.prev = None  # Pointer to the previous song
        self.next = None  # Pointer to the next song

class Playlist:
    def __init__(self):
        self.head = None     
        self.current = None  

    def add_song(self, name):
        new_song = SongNode(name)
        
        if not self.head:
            self.head = new_song
            self.current = self.head
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_song
            new_song.prev = temp  
            
        print(f"Added: {name}")

    def next_song(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"Now playing: {self.current.name}")

    def previous_song(self):
        if self.current and self.current.next:
            self.current = self.current.prev 
            print(f"Now playing: {self.current.name}")
    def display(self):
        temp = self.head
        print("Playlist:", end=" ")
        while temp:
            print(f"{temp.name}", end=" => ")
            temp = temp.next
        print("(None)")

Playlist = Playlist()
Playlist.add_song("Shape of You")
Playlist.add_song("Blinding Lights")
Playlist.add_song("Levitating")
Playlist.display()
print(f"Current Song: {Playlist.current.name}")
Playlist.next_song() 
Playlist.next_song()
Playlist.previous_song() # Should play "Blinding Lights"
```
