```python
class Coach:
    def __init__(self, name):
        self.name = name
        self.prev = None 
        self.next = None  

class Train:
    def __init__(self):
        self.head = None

    def insert_head(self, name):
        """Adds a coach to the head (front) of the train."""
        n = Coach(name)
        n.next = self.head 
        if self.head:
            self.head.prev = n  
        self.head = n 
        
    def insert_tail(self, name):
        """Adds a coach to the tail (end) of the train."""
        n = Coach(name)
        
        if not self.head:
            self.head = n
            return
            
        t = self.head
        while t.next:
            t = t.next
        t.next = n
        n.prev = t
    
    def delete_head(self):
        """Removes a coach from the head of the train."""
        if self.head:
            self.head = self.head.next
            if self.head:
                self.head.prev = None
                
    def delete_tail(self):
        """Removes a coach from the tail of the train."""
        if not self.head: 
            return 
        t = self.head
        if not t.next:
            self.head = None 
            return
        while t.next.next:
            t = t.next
        t.next = None

    def display(self):
        """Prints the sequence of coaches in the train."""
        temp = self.head
        coaches = []
        while temp:
            coaches.append(temp.name)
            temp = temp.next
        print("Train Coaches: " + " <-> ".join(coaches) if coaches else "Train is empty.")

print("--- Initializing Train ---")
train = Train()
train.insert_head("A1") 
train.insert_tail("B1")
train.insert_head("A0") 
train.display()
train.delete_head() 
train.delete_tail() 
train.display()
```
