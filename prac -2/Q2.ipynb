```python
# Smt. Chandibai Himathmal Mansukhani College

class Node:
    def __init__(self, coeff, exp):
        self.coeff = coeff
        self.exp = exp
        self.next = None

class Polynomial:
    def __init__(self):
        self.head = None
    
    def add_term(self, coeff, exp):
        new_node = Node(coeff, exp)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
    
    def input_poly(self):
        n = int(input("Enter the number of terms: "))
        for i in range(n):
            coeff = int(input("Coefficient: "))
            exp = int(input("Exponent: "))
            self.add_term(coeff, exp)
    
    def display(self):
        if self.head is None:
            return "0"
        
        terms = []
        current = self.head
        while current is not None:
            if current.coeff != 0:
                if current.exp == 0:
                    terms.append(f"{current.coeff}")
                elif current.exp == 1:
                    terms.append(f"{current.coeff}x")
                else:
                    terms.append(f"{current.coeff}x^{current.exp}")
            current = current.next
        
        # Sort by exponent in descending order
        terms_sorted = sorted(terms, key=lambda x: int(x.split('^')[1]) if '^' in x else (1 if 'x' in x else 0), reverse=True)
        
        return " + ".join(terms_sorted).replace("+ -", "- ")
    
    @staticmethod
    def add(p1, p2):
        result = Polynomial()
        current1 = p1.head
        current2 = p2.head
        
        while current1 is not None or current2 is not None:
            if current1 is not None and current2 is not None:
                if current1.exp == current2.exp:
                    result.add_term(current1.coeff + current2.coeff, current1.exp)
                    current1 = current1.next
                    current2 = current2.next
                elif current1.exp > current2.exp:
                    result.add_term(current1.coeff, current1.exp)
                    current1 = current1.next
                else:
                    result.add_term(current2.coeff, current2.exp)
                    current2 = current2.next
            elif current1 is not None:
                result.add_term(current1.coeff, current1.exp)
                current1 = current1.next
            else:
                result.add_term(current2.coeff, current2.exp)
                current2 = current2.next
        
        return result
    
    @staticmethod
    def subtract(p1, p2):
        result = Polynomial()
        current1 = p1.head
        current2 = p2.head
        
        while current1 is not None or current2 is not None:
            if current1 is not None and current2 is not None:
                if current1.exp == current2.exp:
                    result.add_term(current1.coeff - current2.coeff, current1.exp)
                    current1 = current1.next
                    current2 = current2.next
                elif current1.exp > current2.exp:
                    result.add_term(current1.coeff, current1.exp)
                    current1 = current1.next
                else:
                    result.add_term(-current2.coeff, current2.exp)
                    current2 = current2.next
            elif current1 is not None:
                result.add_term(current1.coeff, current1.exp)
                current1 = current1.next
            else:
                result.add_term(-current2.coeff, current2.exp)
                current2 = current2.next
        
        return result
    
    @staticmethod
    def multiply(p1, p2):
        result = Polynomial()
        current1 = p1.head
        
        while current1 is not None:
            current2 = p2.head
            while current2 is not None:
                coeff = current1.coeff * current2.coeff
                exp = current1.exp + current2.exp
                
                # Add to result polynomial
                temp = result.head
                found = False
                while temp is not None:
                    if temp.exp == exp:
                        temp.coeff += coeff
                        found = True
                        break
                    temp = temp.next
                
                if not found:
                    result.add_term(coeff, exp)
                
                current2 = current2.next
            current1 = current1.next
        
        return result

# Main program
print("Enter first polynomial:")
p1 = Polynomial()
p1.input_poly()

print("Enter second polynomial:")
p2 = Polynomial()
p2.input_poly()

print("P1 =", p1.display())
print("P2 =", p2.display())

p_add = Polynomial.add(p1, p2)
print("Addition =", p_add.display())

p_sub = Polynomial.subtract(p1, p2)
print("Subtraction =", p_sub.display())

p_mul = Polynomial.multiply(p1, p2)
print("Multiplication =", p_mul.display())
```
