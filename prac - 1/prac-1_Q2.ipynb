```python
# Class representing a Book
class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author

    def __str__(self):
        return f"Book ID: {self.book_id}, Title: {self.title}, Author: {self.author}"


# Class for Book Inventory
class BookInventory:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def delete_book_by_title(self, title):
        self.books = [book for book in self.books if book.title != title]

    def update_book_author(self, book_id, new_author):
        for book in self.books:
            if book.book_id == book_id:
                book.author = new_author
                print(f"Author updated for {book.title}")
                break

    def display_books_sorted_by_title(self):
        sorted_books = sorted(self.books, key=lambda b: b.title)
        print("\nBooks in Inventory (sorted by Title):")
        print(f"{'Book ID':<8} {'Title':<25} {'Author':<15}")
        print("-" * 50)
        for book in sorted_books:
            print(f"{book.book_id:<8} {book.title:<25} {book.author:<15}")


# ----------------- MAIN PROGRAM -----------------
inventory = BookInventory()

# a. Add Books
inventory.add_book(Book("B001", "Python Basics", "John Doe"))
inventory.add_book(Book("B002", "Data Structures", "Jane Smith"))
inventory.add_book(Book("B003", "Machine Learning", "Alan Turing"))

# b. Delete "Data Structures"
inventory.delete_book_by_title("Data Structures")

# c. Add new book B004
inventory.add_book(Book("B004", "Database Systems", "E. Codd"))

# d. Update author of Machine Learning (B003) to "A. Turing"
inventory.update_book_author("B003", "A. Turing")

# e. Sort and Display books by Title
inventory.display_books_sorted_by_title()
```
