```python
class Student:
    def __init__(self, roll_no, name, attendance):
        self.roll_no = roll_no
        self.name = name
        self.attendance = attendance

    def __str__(self):
        return f"Roll No: {self.roll_no}, Name: {self.name}, Attendance: {self.attendance}%"

# Function to display list of students
def display_students(students):
    for s in students:
        print(s)

# ------------------ MAIN PROGRAM ------------------

students = []

# Accept details for 5 students
for i in range(5):
    roll_no = input(f"Enter Roll No of student {i+1}: ")
    name = input(f"Enter Name of student {i+1}: ")
    attendance = float(input(f"Enter Attendance Percentage of student {i+1}: "))
    students.append(Student(roll_no, name, attendance))

# Update attendance for a student
roll_update = input("\nEnter the Roll No to update attendance: ")
for s in students:
    if s.roll_no == roll_update:
        s.attendance = float(input("Enter new attendance percentage: "))
        print("Attendance updated successfully!\n")
        break
else:
    print("Student not found!\n")

# Delete record of a student
roll_delete = input("Enter Roll No of student to delete: ")
students = [s for s in students if s.roll_no != roll_delete]
print("Record deleted successfully!\n")

# Display students with attendance less than 75%
print("Students with attendance less than 75%:")
low_attendance = [s for s in students if s.attendance < 75]
display_students(low_attendance)
print()

# Display students sorted by attendance percentage
print("Students sorted by attendance percentage (ascending):")
students_sorted = sorted(students, key=lambda s: s.attendance)
display_students(students_sorted)
```
