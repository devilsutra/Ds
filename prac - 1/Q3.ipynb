```python
class Book:
    def __init__(self, book_id, title, author):
        self.__book_id = book_id
        self.__title = title
        self.__author = author
    
    def get_book_id(self):
        return self.__book_id
    
    def get_title(self):
        return self.__title
    
    def get_author(self):
        return self.__author
    
    def set_author(self, author):
        self.__author = author
    
    def __str__(self):
        return f"{self.__book_id:10} {self.__title:20} {self.__author:15}"

class BookInventory:
    def __init__(self):
        self.__books = []
    
    def add_book(self, book):
        self.__books.append(book)
    
    def delete_book_by_title(self, title):
        self.__books = [book for book in self.__books if book.get_title() != title]
    
    def update_book_author(self, book_id, new_author):
        for book in self.__books:
            if book.get_book_id() == book_id:
                book.set_author(new_author)
                break
    
    def display_books_sorted_by_title(self):
        sorted_books = sorted(self.__books, key=lambda book: book.get_title())
        print("BookId     Title               Author")
        print("-" * 45)
        for book in sorted_books:
            print(book)

# Main program
inventory = BookInventory()
inventory.add_book(Book("B001", "Python Basics", "John Doe"))
inventory.add_book(Book("B002", "Data Structures", "Jane Smith"))
inventory.add_book(Book("B003", "Machine Learning", "Alan Trung"))

inventory.delete_book_by_title("Data Structures")

inventory.add_book(Book("B004", "Database Systems", "E. Cooks"))

inventory.update_book_author("B003", "A. Turing")

inventory.display_books_sorted_by_title()
```
