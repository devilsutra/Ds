```python
from collections import deque
linear_queue = []

def l_enqueue(item):
    linear_queue.append(item)
    print(f"{item} added to Linear Queue.")

def l_dequeue():
    if not linear_queue:
        print("Linear Queue is empty")
    else:
        served = linear_queue.pop(0)
        print(f"{served} served and removed from Linear Queue.")

def l_show():
    print("Linear Queue:", linear_queue)
circular_queue = deque(maxlen=3) 
circular_queue_maxlen = 3 

def c_enqueue(item):
    if len(circular_queue) == circular_queue_maxlen:
        print(f"Circular Queue full! Oldest element will be overwritten.")
        
    circular_queue.append(item)
    print(f"{item} added to Circular Queue.")

def c_dequeue():
    if not circular_queue:
        print("Circular Queue is empty")
    else:
        task = circular_queue.popleft()
        print(f"{task} executed and removed from Circular Queue.")

def c_show():
    print("Circular Queue:", list(circular_queue))

while True:
    print("\n---------- Queue Menu ----------")
    print("1. Enqueue in Linear Queue")
    print("2. Dequeue from Linear Queue")
    print("3. Show Linear Queue")
    print("4. Enqueue in Circular Queue")
    print("5. Dequeue from Circular Queue")
    print("6. Show Circular Queue")
    print("7. Exit")

    choice = input('Enter your choice: ')

    if choice == '1':
        item = input("Enter task name: ")
        l_enqueue(item)
        
    elif choice == '2':
        l_dequeue()
        
    elif choice == '3':
        l_show()
        
    elif choice == '4':
        item = input("Enter task name: ")
        c_enqueue(item)
        
    elif choice == '5':
        c_dequeue()
        
    elif choice == '6':
        c_show()
        
    elif choice == '7':
        print('Exiting Program...')
        break 
        
    else:
        print('Invalid choice! Try again.')
```
